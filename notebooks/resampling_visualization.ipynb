```python
import matplotlib.pyplot as plt
import nibabel as nib
import sys
sys.path.append("..")

from mpl_toolkits.mplot3d import Axes3D
from resample_trk import fiber_geometry

import warnings
warnings.filterwarnings('ignore')
```

```python
trk_file = nib.streamlines.load("../subjects/992774/tracts/CC.trk")
```

```python
def show(fiber, smoothing=5, npts="auto"):
    
    r, _, _, _, _, _, _ = fiber_geometry(fiber, npts=npts, smoothing=smoothing)
    
    fig2 = plt.figure(dpi=200)

    ax3d = fig2.add_subplot(111, projection='3d')
    ax3d.plot(fiber[:,0], fiber[:,1], fiber[:,2], 'b')
    ax3d.plot(r[:,0], r[:,1], r[:,2], 'xg')
    fig2.show()
    plt.show()
```

```python
show(trk_file.tractogram.streamlines[0])
```
